Importing Dependecies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error,r2_score
```

Importing the Dataset into Pandas DataFrame

```python
# Reading the Dataset
df=pd.read_csv("/content/height-weight.csv")
```

```python
# Printing the Dataset in the Pandas DataFrame
df.head()
```

```python
# Checking for Null Values
df.isnull().sum()
```

```python
# Checking how the data is spread
plt.scatter(df["Height"],df["Weight"])
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python

# Finding the Correlation
df.corr()
```

```python
# Seaborn for Visualization
import seaborn as sns
sns.pairplot(df)
```

```python
x=df[["Height"]].values
y=df["Weight"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
# Visualizing the Training Results
plt.scatter(x_train, y_train, color='red')  # Scatter plot of training data
plt.plot(x_train, regressor.predict(x_train), color='blue')  # Regression line

# Adding labels and title
plt.xlabel('Height (cm)')  # X-axis label for Height
plt.ylabel('Weight (kg)')  # Y-axis label for Weight
plt.title('Height vs Weight Prediction (Training Set)')  # Optional title

plt.show()


```

```python
# Visualizing the Testing Results
plt.scatter(x_test, y_test, color='red')  # Scatter plot of testing data
plt.plot(x_train, regressor.predict(x_train), color='blue')  # Regression line based on training data

# Adding labels and title
plt.xlabel('Height (cm)')  # X-axis label for Height
plt.ylabel('Weight (kg)')  # Y-axis label for Weight
plt.title('Height vs Weight Prediction (Testing Set)')  # Optional title

plt.show()
```

```python

# Accuracy Score Calculation
from sklearn.metrics import r2_score
print(r2_score(y_pred,y_test))
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```
